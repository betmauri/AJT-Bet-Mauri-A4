```python
from IPython.display import clear_output

def mostra_taulell(taulell, diccionari_jugadors, n):
    print("    " + "   ".join(str(i) for i in range(n)))
    for fila in range(n):
        contingut = " | ".join(diccionari_jugadors[taulell[fila][columna]] for columna in range(n))
        print(f" {fila}  {contingut}")
        if fila < n-1:
            print("   " + "---+" *(n-1) + "---")
    print()
    return


def demana_posicio(taulell, n):
    try:
        fila = int(input("Dona'm la fila on vols posar la peça (valor entre 0 i n-1)"))
        while fila < 0 or fila > n-1:
            fila = int(input("Siusplau, dona'm be la fila on vols posar la peça (valor entre 0 i n-1)"))
    except:
        return None, None
    try:
        columna = int(input("Dona'm la columna on vols posar la peça (valor entre 0 i n-1)"))
        while columna < 0 or columna > n-1:
            columna = int(input("Siusplau, dona'm be la fila on vols posar la peça (valor entre 0 i n-1)"))
    except:
        return None, None
    

    while taulell[fila][columna] != 0:
        fila = int(input("Introdueix una posicio que no estigui ocupada"))
        columna = int(input("Introdueix una posicio que no estigui ocupada"))

    return fila, columna

def actualitza_taulell(taulell, fila, columna, jugador):
    taulell[fila][columna] = jugador
    return taulell

def canvi_jugador(jugador):
    return jugador*(-1)

def comprovar_victoria(taulell,jugador, n):
    # Victòria per files
    for fila in range(0,n):
        suma = 0
        for columna in range(0,n):
            suma += taulell[fila][columna]
        if suma == n*jugador:
            return True
    # Victòria per columnes
    for columna in range(0,n):
        suma = 0
        for fila in range(0,n):
            suma += taulell[fila][columna]

        if suma == n*jugador:
            return True 
    # Victòria per diagonals
    suma = 0
    for diagonal in range(0,n):
        suma += taulell[diagonal][diagonal]            
    if suma == n*jugador:
        return True 
    # Victòria per diagonal
    suma = 0
    for diagonal in range(0,n):
        suma += taulell[diagonal][2-diagonal]            
    if suma == n*jugador:
        return True 
    
    return False
```

```python
jugador = 1
diccionari_jugadors = {
    0: " ",
    1: "X",
    -1: "O"
    }

n = int(input("Introdueix la mida que vols del taulell"))
taulell= [[0]*n for i in range(n)]
condicio_victoria = False
comptador = 0


while not condicio_victoria and comptador < n**2:
    clear_output(wait=True)
    mostra_taulell(taulell,diccionari_jugadors, n)
    fila, columna = demana_posicio(taulell, n)
    if fila == None and columna == None:
        print("Has de donar un nombre natural entre 0 i n-1. Sortim del joc")
        break
    taulell = actualitza_taulell(taulell, fila, columna, jugador)
    condicio_victoria = comprovar_victoria(taulell, jugador, n)
    if condicio_victoria:
        clear_output(wait=True)
        mostra_taulell(taulell,diccionari_jugadors, n)
        print(f"Ha guanyat el jugador {diccionari_jugadors[jugador]}")
        break

    comptador += 1
    jugador = canvi_jugador(jugador)
```
