```python
#1
import math
def concentracio(concentracio_inicial, k, t):
    concentraciofinal = concentracio_inicial*math.exp(-k*t)
    return concentraciofinal

for T in range(0, 26):
    c = concentracio(200, 3, T)
    print(c)
```

```python
#3
def fibonacci(n):
    if n ==0:
        F = [0]
        return F
    elif n == 1:
        F = [0, 1]
        return F
    else:
        F = [0, 1]
        for i in range(0, n):
            a = F[-2] + F[-1]
            F.append(a)
    return F



fibonacci(30)

```
